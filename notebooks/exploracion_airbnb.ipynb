```python
import sys
import os

# Esto le dice al Notebook que busque los scripts en la carpeta src
sys.path.append(os.path.abspath(os.path.join('..', 'src')))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from extraccion import Extraccion

%matplotlib inline
pd.set_option('display.max_columns', None)

extractor = Extraccion("mongodb://localhost:27017/", "AirbnbBuenosAires")
datos = extractor.ejecutar_extraccion_completa()

df_listings = datos['listings']
df_calendar = datos['calendar']
df_reviews = datos['reviews']

print("--- LISTINGS ---")
display(df_listings.head())
print(f"Registros: {df_listings.shape[0]} | Columnas: {df_listings.shape[1]}")
df_listings.info()

print("--- CALENDAR ---")
display(df_calendar.head())
print(f"Registros: {df_calendar.shape[0]} | Columnas: {df_calendar.shape[1]}")
df_calendar.info()

print("\n--- REVIEWS ---")
display(df_reviews.head())
print(f"Registros: {df_reviews.shape[0]} | Columnas: {df_reviews.shape[1]}")
df_reviews.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

print("=== NULOS ===")
nulos_columna = df_listings.isnull().sum().sort_values(ascending=False).head(15)
print(nulos_columna)

print("\n=== DUPLICADOS ===")
duplicados_id = df_listings.duplicated(subset=['id']).sum()
print(f"Registros con ID duplicado en Listings: {duplicados_id}")
print(f"Registros duplicados en Reviews: {df_reviews.duplicated().sum()}")

print("\n=== OUTLIERS (VALORES ATÍPICOS) ===")
sns.set_theme(style="whitegrid")

variables_outliers = ['minimum_nights', 'availability_365', 'accommodates']

plt.figure(figsize=(16, 5))
for i, var in enumerate(variables_outliers):
    plt.subplot(1, 3, i+1)
    sns.boxplot(y=df_listings[var], color='salmon')
    plt.title(f'Distribución de {var}')
    plt.ylabel('Valor')

plt.tight_layout()
plt.show()
```

```python
print("Iniciando proceso de transformación...")

# Estandarización de FECHAS
df_listings['last_review'] = pd.to_datetime(df_listings['last_review'])
df_calendar['date'] = pd.to_datetime(df_calendar['date'])
df_reviews['date'] = pd.to_datetime(df_reviews['date'])

# Transformación de PRECIO
if 'price' in df_calendar.columns:
    df_calendar['price'] = df_calendar['price'].replace(r'[\$,]', '', regex=True).astype(float)
    print("Columna 'price' estandarizada.")
else:
    print("Variable 'price' no detectada. Se omite su conversión.")

# Calculamos la ocupación estimada para los próximos 30 y 90 días
df_listings['occupancy_30_pct'] = ((30 - df_listings['availability_30']) / 30) * 100
df_listings['occupancy_90_pct'] = ((90 - df_listings['availability_90']) / 90) * 100

# Segmentamos por tipo de estancia basándonos en el mínimo de noches
df_listings['rental_category'] = df_listings['minimum_nights'].apply(
    lambda x: 'Larga Estancia (>30d)' if x > 30 else ('Media Estancia' if x > 7 else 'Turístico Corto')
)

# Conteo de amenidades (desanidado simple)
df_listings['amenities_count'] = df_listings['amenities'].apply(lambda x: len(x) if isinstance(x, list) else 0)

# Limpieza de nombres de barrios
df_listings['neighbourhood_cleansed'] = df_listings['neighbourhood_cleansed'].str.strip().str.title()

# 5. TEMPORALIDAD EN CALENDAR
df_calendar['month'] = df_calendar['date'].dt.month
df_calendar['year'] = df_calendar['date'].dt.year

print("\n=== TRANSFORMACIONES FINALIZADAS ===")
# Vista previa de las nuevas variables de BI
display(df_listings[['name', 'neighbourhood_cleansed', 'occupancy_30_pct', 'rental_category']].head())
```

```python
# Top 10 barrios con mayor ocupación
plt.figure(figsize=(10, 6))
df_listings.groupby('neighbourhood_cleansed')['occupancy_30_pct'].mean().sort_values(ascending=False).head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Barrios con Mayor Ocupación en los próximos 30 días')
plt.ylabel('% de Ocupación Promedio')
plt.xticks(rotation=45)
plt.show()
```
